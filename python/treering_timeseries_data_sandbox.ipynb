```python
import json
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline
```

```python
contents = open("C:/Users/Jacob/Projects/itrdb/itrdb_rwl_data.json")
contents = contents.read()
data = json.loads(contents)
```

```python
data.keys()
```

```python
plt.figure(figsize=(15,6))
count = 0
for t in data['wv003.rwl']:
    x = data['wv003.rwl'][t][1][1:-1]
    y = np.array(data['wv003.rwl'][t][0][1:-1])
    s = UnivariateSpline(x, y, s=1)
    xs = np.linspace(data['wv003.rwl'][t][1][1],data['wv003.rwl'][t][1][-2], 1000)
    ys = s(xs)
    plt.plot(x, y, '.-')
    plt.plot(xs, ys)
    count+=1
    if count == 4:
        break
plt.show()
```

```python
from numpy import linspace,exp
from numpy.random import randn
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
x = linspace(-3, 3, 100)
y = exp(-x**2) + randn(100)/10
s = UnivariateSpline(x, y, s=1)
xs = linspace(-3, 3, 1000)
ys = s(xs)
plt.plot(x, y, '.-')
plt.plot(xs, ys)
plt.show()
```
